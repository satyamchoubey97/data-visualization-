```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=df2, palette='Set3')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x labels if needed for better readability
plt.legend(title='Attrition')  # Add legend title
plt.show()
```

```python
for department in attrition_counts.index:
    plt.fill_between(attrition_counts.columns, attrition_counts.loc[department], label=department, alpha=0.5)

```

```python
plt.title('Attrition Distribution by Department (Area Chart)')
plt.xlabel('Attrition Status')
plt.ylabel('Count')
plt.xticks(ticks=range(len(attrition_counts.columns)), labels=attrition_counts.columns, rotation=45)
plt.legend(title='Department')
plt.grid()
plt.show()
```
